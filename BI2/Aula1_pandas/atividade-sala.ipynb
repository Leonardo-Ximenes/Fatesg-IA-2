```python
import pandas as pd
```

```python
data = {'Name': ['João', 'Maria', 'Pedro', 'Anna'], 'Age' :[25,30,35,20]}
df = pd.DataFrame(data)
df
```

```python
df = pd.DataFrame(data, index = ['Rank1', 'Rank2', 'Rank3', 'Rank4'])
df 
```

```python
data = [{'a': 1, 'b': 2}, {'a': 1, 'b': 2, 'c': 3}]
df = pd.DataFrame(data)
df
#Vai dar erro (NaN), porque uma parte tem 2 elementos e a outra tem 3
```

```python
df1 = pd.DataFrame(data, index =['first', 'second'], columns= ['a', 'b'])

df2= pd.DataFrame(data, index =['first', 'second'], columns= ['a', 'b1'])

print (df1)
print (df2)
```

```python
dados = {'one' : pd.Series([1,2,3], index = ['a', 'b', 'c']), 'two' : pd.Series([1,2,3,4], index = ['a', 'b', 'c', 'd'])}
df = pd.DataFrame(dados)
df 
```

```python
print ("Adicionando uma nova coluna passando como Series")
df['three'] = pd.Series([10,20,30], index = ['a', 'b', 'c'])
print (df)

print ("Adicionando uma nova coluna usando as colunas existentes no DataFrame:")
df['four']=df['one']+df['three']

print (df)
```

```python
df = pd.DataFrame([[1,2], [3,4]], columns = ['a','b'])
df2 = pd.DataFrame([[5,6], [7,8]], columns = ['a','b'])
df = pd.concat([df, df2])

#Excluindo as linhas com rótulo 0
df = df.drop(0)

print (df)
```
